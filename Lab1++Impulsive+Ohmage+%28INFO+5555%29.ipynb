```python
delay=pandas.read_csv('/Users/arpitsabherwal/Downloads/delaydiscounting.csv')
```

```python
#Parsing Data
delay["Now Array"]=delay["Now Array"].str[1:]
delay["Unnamed: 8"]=delay["Unnamed: 8"].str[:-1]
delay["Later Array"]=delay["Now Array"].str[1:]
delay["Unnamed: 14"]=delay["Unnamed: 14"].str[:-1]
delay["Choice Array"]=delay["Choice Array"].str[1:]
delay["Unnamed: 20"]=delay["Unnamed: 20"].str[:-1]
delay["Response Times"]=delay["Response Times"].str[1:]
delay["Unnamed: 26"]=delay["Unnamed: 26"].str[:-1]
```

```python
delay["Now Array"]=delay["Now Array"].convert_objects(convert_numeric=True)
delay["Unnamed: 8"]=delay["Unnamed: 8"].convert_objects(convert_numeric=True)
delay["Later Array"]=delay["Now Array"].convert_objects(convert_numeric=True)
delay["Unnamed: 14"]=delay["Unnamed: 14"].convert_objects(convert_numeric=True)
delay["Choice Array"]=delay["Choice Array"].convert_objects(convert_numeric=True)
delay["Unnamed: 20"]=delay["Unnamed: 20"].convert_objects(convert_numeric=True)
delay["Response Times"]=delay["Response Times"].convert_objects(convert_numeric=True)
delay["Unnamed: 26"]=delay["Unnamed: 26"].convert_objects(convert_numeric=True)
```

```python
plt.plot(np.arange(6),now_array[0])
plt.plot(np.arange(6),later_array[0])
plt.plot(np.arange(6),response_times[0]*10)
plt.legend(['now_array','later_array','response_times'],loc='upper right')
plt.xlabel("Choice Number")
plt.show()
```
